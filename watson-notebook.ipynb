## Preparing Data for Input

To start out, we can use a pretrained model. Here, we'll use a multilingual BERT model from huggingface. For more information about BERT, see: https://github.com/google-research/bert/blob/master/multilingual.md

First, we download the tokenizer.

```python
model_name = 'bert-base-multilingual-cased'
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

Tokenizers turn sequences of words into arrays of numbers. Let's look at an example:

```python
def encode_sentence(s):
    return tokenizer.encode(s, add_special_tokens=False) + [tokenizer.sep_token_id]
```

```python
encode_sentence("I love machine learning")
```

BERT uses three kind of input data- input word IDs, input masks, and input type IDs.

These allow the model to know that the premise and hypothesis are distinct sentences, and also to ignore any padding from the tokenizer.

We add a [CLS] token to denote the beginning of the inputs, and a [SEP] token to denote the separation between the premise and the hypothesis. We also need to pad all of the inputs to be the same size. For more information about BERT inputs, see: https://huggingface.co/transformers/model_doc/bert.html#bertmodel

Now, we're going to encode all of our premise/hypothesis pairs for input into BERT.

`max_len` was originally 50, but with the mDeBERTa tokenizer that truncates 45.8% of premise+hypothesis pairs (median full length: 48 tokens) — often cutting off the hypothesis entirely, which is half the information the model needs to classify entailment/neutral/contradiction. Raised to 128, which covers 99.4% of examples untruncated. This is a shared preprocessing constant used by both the baseline and mDeBERTa sections, so it's fixed for both to keep the comparison fair.

```python
max_len = 128  # con el tokenizer de mDeBERTa, max_len=50 truncaba el 45.8% de los ejemplos (mediana premise+hypothesis: 48 tokens); 128 cubre el 99.4%

def bert_encode(premises, hypotheses, tokenizer, max_len=max_len):
    encodings = tokenizer(
        list(premises), list(hypotheses),
        padding='max_length', truncation=True, max_length=max_len,
        return_tensors='pt')
    return {
        'input_ids': encodings['input_ids'],
        'attention_mask': encodings['attention_mask'],
        'token_type_ids': encodings['token_type_ids'],
    }
```

```python
train_input = bert_encode(train.premise.values, train.hypothesis.values, tokenizer)
```

```python
from sklearn.model_selection import train_test_split

class NLIDataset(Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels

    def __len__(self):
        return len(self.labels)

    def __getitem__(self, idx):
        item = {k: v[idx] for k, v in self.encodings.items()}
        item['label'] = self.labels[idx]
        return item

labels = torch.tensor(train.label.values, dtype=torch.long)
train_idx, val_idx = train_test_split(
    np.arange(len(labels)), test_size=0.2, random_state=42)

train_encodings = {k: v[train_idx] for k, v in train_input.items()}
val_encodings = {k: v[val_idx] for k, v in train_input.items()}

train_dataset = NLIDataset(train_encodings, labels[train_idx])
val_dataset = NLIDataset(val_encodings, labels[val_idx])

batch_size = 32
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=batch_size)
```

## Creating & Training Model

Now, we can incorporate the BERT transformer into a PyTorch model. This model was inspired by the model in this notebook: https://www.kaggle.com/tanulsingh077/deep-learning-for-nlp-zero-to-transformers-bert#BERT-and-Its-Implementation-on-this-Competition, which is a wonderful introduction to NLP!

```python
class NLIModel(nn.Module):
    def __init__(self, model_name, num_labels=3, dropout=0.0, pooling='cls'):
        super().__init__()
        self.bert = AutoModel.from_pretrained(model_name)
        self.dropout = nn.Dropout(dropout)
        self.classifier = nn.Linear(self.bert.config.hidden_size, num_labels)
        self.pooling = pooling

    def forward(self, input_ids, attention_mask, token_type_ids):
        outputs = self.bert(
            input_ids=input_ids,
            attention_mask=attention_mask,
            token_type_ids=token_type_ids)
        hidden_states = outputs.last_hidden_state
        if self.pooling == 'mean':
            mask = attention_mask.unsqueeze(-1).type_as(hidden_states)
            pooled = (hidden_states * mask).sum(dim=1) / mask.sum(dim=1).clamp(min=1e-9)
        else:
            pooled = hidden_states[:, 0, :]
        pooled = self.dropout(pooled)
        return self.classifier(pooled)
```

```python
model = NLIModel(model_name).to(device)
num_params = sum(p.numel() for p in model.parameters())
print(f'Total parameters: {num_params:,}')
```

We also checkpoint the model weights after every epoch — `baseline_last.pt` is overwritten each epoch, while `baseline_best.pt` is only updated when validation loss improves — and apply early stopping if validation loss hasn't improved for 5 consecutive epochs.

```python
checkpoint_dir = 'checkpoints'
os.makedirs(checkpoint_dir, exist_ok=True)

optimizer = torch.optim.AdamW(model.parameters(), lr=1e-5)
criterion = nn.CrossEntropyLoss()

epochs = 2
patience = 5
best_val_loss = float('inf')
baseline_best_val_accuracy = None
epochs_no_improve = 0

for epoch in range(epochs):
    model.train()
    train_loss, train_correct, train_total = 0.0, 0, 0
    for batch in train_loader:
        optimizer.zero_grad()
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        token_type_ids = batch['token_type_ids'].to(device)
        batch_labels = batch['label'].to(device)

        logits = model(input_ids, attention_mask, token_type_ids)
        loss = criterion(logits, batch_labels)
        loss.backward()
        optimizer.step()

        train_loss += loss.item() * batch_labels.size(0)
        train_correct += (logits.argmax(dim=-1) == batch_labels).sum().item()
        train_total += batch_labels.size(0)

    model.eval()
    val_loss, val_correct, val_total = 0.0, 0, 0
    with torch.no_grad():
        for batch in val_loader:
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            token_type_ids = batch['token_type_ids'].to(device)
            batch_labels = batch['label'].to(device)

            logits = model(input_ids, attention_mask, token_type_ids)
            loss = criterion(logits, batch_labels)

            val_loss += loss.item() * batch_labels.size(0)
            val_correct += (logits.argmax(dim=-1) == batch_labels).sum().item()
            val_total += batch_labels.size(0)

    baseline_val_accuracy = val_correct / val_total
    avg_val_loss = val_loss / val_total
    print(f'Epoch {epoch+1}/{epochs} - '
          f'loss: {train_loss/train_total:.4f} - accuracy: {train_correct/train_total:.4f} - '
          f'val_loss: {avg_val_loss:.4f} - val_accuracy: {baseline_val_accuracy:.4f}')

    # Checkpointing: "last" is overwritten every epoch, "best" only when val_loss improves
    torch.save(model.state_dict(), os.path.join(checkpoint_dir, 'baseline_last.pt'))

    if avg_val_loss < best_val_loss:
        best_val_loss = avg_val_loss
        baseline_best_val_accuracy = baseline_val_accuracy
        epochs_no_improve = 0
        torch.save(model.state_dict(), os.path.join(checkpoint_dir, 'baseline_best.pt'))
        print(f'  -> new best val_loss ({best_val_loss:.4f}), saved to baseline_best.pt')
    else:
        epochs_no_improve += 1
        print(f'  -> val_loss did not improve for {epochs_no_improve} epoch(s) (best: {best_val_loss:.4f})')
        if epochs_no_improve >= patience:
            print(f'Early stopping: no val_loss improvement for {patience} consecutive epochs.')
            break
```

Since this is a 3-class classification problem (entailment / neutral / contradiction), accuracy alone hides *which* classes get confused with each other. We reload the best checkpoint (lowest val_loss, not necessarily the last epoch trained) and plot the confusion matrix on the validation split.

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

class_names = ['entailment', 'neutral', 'contradiction']

model.load_state_dict(torch.load(os.path.join(checkpoint_dir, 'baseline_best.pt')))
model.eval()

val_preds, val_true = [], []
with torch.no_grad():
    for batch in val_loader:
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        token_type_ids = batch['token_type_ids'].to(device)
        batch_labels = batch['label'].to(device)

        logits = model(input_ids, attention_mask, token_type_ids)
        val_preds.extend(logits.argmax(dim=-1).cpu().tolist())
        val_true.extend(batch_labels.cpu().tolist())

cm = confusion_matrix(val_true, val_preds)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=class_names)
fig, ax = plt.subplots(figsize=(6, 6))
disp.plot(ax=ax, cmap='Blues', colorbar=False)
ax.set_title('Baseline (mBERT) - confusion matrix (val, best checkpoint)')
plt.show()
```

## Mejora: transfer learning desde un modelo preentrenado en NLI (mDeBERTa)

El baseline de arriba parte de `bert-base-multilingual-cased`, preentrenado solo con masked-LM genérico — no sabe nada de NLI de partida, así que tiene que aprender la tarea entera durante el fine-tuning.

Aquí partimos en cambio de [`MoritzLaurer/mDeBERTa-v3-base-mnli-xnli`](https://huggingface.co/MoritzLaurer/mDeBERTa-v3-base-mnli-xnli), que ya está fine-tuneado en MNLI + XNLI (mismo formato premise/hypothesis/entailment-neutral-contradiction que esta competición, en 100 idiomas). Congelamos las capas inferiores del encoder (features lingüísticas más genéricas, ya bien aprendidas) y solo entrenamos las últimas capas más un classifier head nuevo — transfer learning real en vez de fine-tuning completo desde cero.

```python
model_name_nli = 'MoritzLaurer/mDeBERTa-v3-base-mnli-xnli'
tokenizer_nli = AutoTokenizer.from_pretrained(model_name_nli)
model_nli = NLIModel(model_name_nli, dropout=0.4, pooling='mean').float().to(device)  # el checkpoint se guardo en fp16; el classifier nuevo es fp32

total_params_nli = sum(p.numel() for p in model_nli.parameters())
```

Congelamos embeddings + capas inferiores del encoder; dejamos entrenables solo las últimas `K` capas del encoder y el classifier head nuevo. El naming de las capas de DeBERTa-v3 (`encoder.layer.{i}.*`) se verificó antes de escribir este filtro.

`K` empezó en 3. Ahora que `max_len=128` arregló el cuello de botella real (el truncamiento), probamos si más capacidad entrenable ayuda: subimos `K` a 6 (la mitad del encoder, 42.5M parámetros entrenables en vez de 21.3M) manteniendo el resto igual para aislar el efecto de este único cambio. Resultado: **el mismo techo exacto de val_accuracy (88.04%)**, pero alcanzado en la época 1 en vez de la 2, con overfitting posterior mucho más agresivo (train_accuracy 97.8% en la época 6) y un coste de entrenamiento mucho mayor (GPU al 100%, VRAM casi al límite, bastante más lento por época). Sin ninguna ganancia a cambio de ese coste, así que volvemos a `K=3`.

```python
K = 3  # nº de capas superiores del encoder que se entrenan (de 12 en total)

for name, param in model_nli.bert.named_parameters():
    if name.startswith('encoder.layer.'):
        layer_idx = int(name.split('.')[2])
        param.requires_grad = layer_idx >= (model_nli.bert.config.num_hidden_layers - K)
    else:
        param.requires_grad = False

for param in model_nli.classifier.parameters():
    param.requires_grad = True

trainable_params_nli = sum(p.numel() for p in model_nli.parameters() if p.requires_grad)
print(f'Parametros entrenables: {trainable_params_nli:,} de {total_params_nli:,} '
      f'({trainable_params_nli/total_params_nli:.1%})')
```

Volvemos a tokenizar premise/hypothesis con el tokenizer de mDeBERTa (vocabulario distinto al de mBERT) reutilizando el mismo split train/val (`train_idx`/`val_idx`) que el baseline, para que la comparación de métricas sea justa.

```python
train_input_nli = bert_encode(train.premise.values, train.hypothesis.values, tokenizer_nli)

train_encodings_nli = {k: v[train_idx] for k, v in train_input_nli.items()}
val_encodings_nli = {k: v[val_idx] for k, v in train_input_nli.items()}

train_dataset_nli = NLIDataset(train_encodings_nli, labels[train_idx])
val_dataset_nli = NLIDataset(val_encodings_nli, labels[val_idx])

train_loader_nli = DataLoader(train_dataset_nli, batch_size=batch_size, shuffle=True)
val_loader_nli = DataLoader(val_dataset_nli, batch_size=batch_size)
```

Entrenamos solo los parámetros no congelados, con warmup + decaimiento lineal del learning rate y gradient clipping — prácticas estándar de fine-tuning de BERT que no estaban en el baseline. Igual que en el baseline, guardamos los pesos en cada epoch (`nli_last.pt` se sobrescribe siempre; `nli_best.pt` solo cuando mejora la val_loss) y aplicamos early stopping si la val_loss no mejora durante 5 epochs consecutivos — con hasta 15 epochs disponibles, esto evita seguir entrenando una vez el modelo converge.

En la primera ejecución se vio overfitting claro: train_accuracy siguió subiendo (hasta ~89%) mientras la val_loss empeoraba a partir de la época 2, así que el early stopping se quedó con esos pesos (~80.4% val_accuracy). Subir la regularización del classifier head (dropout `p=0.2→0.4`, `weight_decay` de 0.01 a 0.1) dio una mejora pequeña y acumulativa (~+0.25 puntos) pero el gap train/val apenas se movió — indicio de que el cuello de botella no está solo en el classifier nuevo, sino en cómo se agrega la representación de la secuencia.

Dos cambios más en esta ejecución: (1) semilla fija (`SEED = 42`, celda al principio del notebook) para que las comparaciones dejen de estar contaminadas por el ~1.3 puntos de ruido que vimos entre runs idénticos del baseline; (2) mean pooling en vez de usar solo el token `[CLS]` — se promedia `last_hidden_state` sobre todos los tokens no-padding (usando `attention_mask`), lo que aprovecha toda la secuencia en vez de un único vector. El baseline no se toca (sigue con `pooling='cls'` por defecto en `NLIModel`).

**Nota sobre esta ejecución concreta:** con la semilla fija, la última corrida completa (15 epochs, early stopping en la época 7) dio su mejor `val_loss` en la época 2 y no volvió a mejorar después. Para esta re-ejecución (pensada solo para regenerar el checkpoint y alimentar el análisis de accuracy por idioma más abajo) fijamos `epochs_nli = 2` en vez de 15 — nos ahorramos las 5 épocas que el patience necesitaba agotar sin cambiar qué pesos termina usando `nli_best.pt`. Si en el futuro se quiere volver a explorar si hay una época mejor más adelante (p. ej. tras cambiar algún hiperparámetro), hay que restaurar `epochs_nli = 15` para que el early stopping tenga margen real de actuar.

```python
from transformers import get_linear_schedule_with_warmup

checkpoint_dir = 'checkpoints'
os.makedirs(checkpoint_dir, exist_ok=True)

optimizer_nli = torch.optim.AdamW(
    filter(lambda p: p.requires_grad, model_nli.parameters()), lr=2e-5, weight_decay=0.1)

epochs_nli = 2  # la ultima ejecucion documentada (SEED=42 fijo) dio su mejor val_loss en la epoca 2; reentrenamos solo hasta ahi en vez de esperar a que el early stopping agote el patience en la epoca 7
patience_nli = 5
total_steps_nli = len(train_loader_nli) * epochs_nli
scheduler_nli = get_linear_schedule_with_warmup(
    optimizer_nli,
    num_warmup_steps=int(0.1 * total_steps_nli),
    num_training_steps=total_steps_nli)

best_val_loss_nli = float('inf')
nli_best_val_accuracy = None
epochs_no_improve_nli = 0

for epoch in range(epochs_nli):
    model_nli.train()
    train_loss, train_correct, train_total = 0.0, 0, 0
    for batch in train_loader_nli:
        optimizer_nli.zero_grad()
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        token_type_ids = batch['token_type_ids'].to(device)
        batch_labels = batch['label'].to(device)

        logits = model_nli(input_ids, attention_mask, token_type_ids)
        loss = criterion(logits, batch_labels)
        loss.backward()
        torch.nn.utils.clip_grad_norm_(model_nli.parameters(), max_norm=1.0)
        optimizer_nli.step()
        scheduler_nli.step()

        train_loss += loss.item() * batch_labels.size(0)
        train_correct += (logits.argmax(dim=-1) == batch_labels).sum().item()
        train_total += batch_labels.size(0)

    model_nli.eval()
    val_loss, val_correct, val_total = 0.0, 0, 0
    with torch.no_grad():
        for batch in val_loader_nli:
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            token_type_ids = batch['token_type_ids'].to(device)
            batch_labels = batch['label'].to(device)

            logits = model_nli(input_ids, attention_mask, token_type_ids)
            loss = criterion(logits, batch_labels)

            val_loss += loss.item() * batch_labels.size(0)
            val_correct += (logits.argmax(dim=-1) == batch_labels).sum().item()
            val_total += batch_labels.size(0)

    nli_val_accuracy = val_correct / val_total
    avg_val_loss_nli = val_loss / val_total
    print(f'Epoch {epoch+1}/{epochs_nli} - '
          f'loss: {train_loss/train_total:.4f} - accuracy: {train_correct/train_total:.4f} - '
          f'val_loss: {avg_val_loss_nli:.4f} - val_accuracy: {nli_val_accuracy:.4f}')

    # Checkpointing: "last" se sobrescribe cada epoch, "best" solo cuando mejora la val_loss
    torch.save(model_nli.state_dict(), os.path.join(checkpoint_dir, 'nli_last.pt'))

    if avg_val_loss_nli < best_val_loss_nli:
        best_val_loss_nli = avg_val_loss_nli
        nli_best_val_accuracy = nli_val_accuracy
        epochs_no_improve_nli = 0
        torch.save(model_nli.state_dict(), os.path.join(checkpoint_dir, 'nli_best.pt'))
        print(f'  -> nuevo mejor val_loss ({best_val_loss_nli:.4f}), guardado en nli_best.pt')
    else:
        epochs_no_improve_nli += 1
        print(f'  -> val_loss sin mejora desde hace {epochs_no_improve_nli} epoch(s) (mejor: {best_val_loss_nli:.4f})')
        if epochs_no_improve_nli >= patience_nli:
            print(f'Early stopping: sin mejora en val_loss durante {patience_nli} epochs consecutivos.')
            break
```

Al ser un problema de clasificación de 3 clases (entailment / neutral / contradiction), la accuracy sola no muestra qué clases se confunden entre sí. Recargamos el checkpoint con mejor val_loss (`nli_best.pt`, no necesariamente el de la última época entrenada) y pintamos la matriz de confusión sobre el split de validación.

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

class_names = ['entailment', 'neutral', 'contradiction']

model_nli.load_state_dict(torch.load(os.path.join(checkpoint_dir, 'nli_best.pt')))
model_nli.eval()

val_preds_nli, val_true_nli = [], []
with torch.no_grad():
    for batch in val_loader_nli:
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        token_type_ids = batch['token_type_ids'].to(device)
        batch_labels = batch['label'].to(device)

        logits = model_nli(input_ids, attention_mask, token_type_ids)
        val_preds_nli.extend(logits.argmax(dim=-1).cpu().tolist())
        val_true_nli.extend(batch_labels.cpu().tolist())

cm_nli = confusion_matrix(val_true_nli, val_preds_nli)
disp_nli = ConfusionMatrixDisplay(confusion_matrix=cm_nli, display_labels=class_names)
fig, ax = plt.subplots(figsize=(6, 6))
disp_nli.plot(ax=ax, cmap='Blues', colorbar=False)
ax.set_title('mDeBERTa (transfer learning) - confusion matrix (val, best checkpoint)')
plt.show()
```

### Accuracy por idioma (mDeBERTa) — dónde augmentar con traducción

La accuracy global (88.0%) puede esconder que el modelo rinda bastante peor en algunos de los 15 idiomas del dataset. Reutilizamos `val_true_nli`/`val_preds_nli` (ya calculados con el mejor checkpoint) y los cruzamos con `train.language.values[val_idx]` para ver la accuracy idioma a idioma en el split de validación.

Esto es sobre el split de **validación**, no sobre `test.csv`: `test.csv` no trae la columna `label` (es el fichero de submission de Kaggle, sin ground truth), así que no hay forma de medir accuracy por idioma ahí directamente — el proxy fiable es el split de validación, que sí tiene etiquetas y viene del mismo `train.csv`.

Los idiomas por debajo de la accuracy global (en rojo) son los candidatos naturales para aumentar datos vía traducción: traducir premise/hypothesis de idiomas donde el modelo ya rinde bien hacia el idioma problemático, manteniendo la label original (la relación de entailment no cambia al traducir). Ojo con el tamaño de muestra: cada idioma no-inglés tiene ~350-400 filas en `train.csv` (~70-80 en validación), así que una accuracy individual con n bajo puede moverse varios puntos solo por ruido de muestreo.

```python
val_lang_nli = train.language.values[val_idx]

lang_df = pd.DataFrame({'language': val_lang_nli, 'true': val_true_nli, 'pred': val_preds_nli})
lang_df['correct'] = lang_df['true'] == lang_df['pred']

overall_acc_nli = lang_df['correct'].mean()
lang_acc = lang_df.groupby('language')['correct'].agg(accuracy='mean', n='size').sort_values('accuracy')
lang_acc
```

## Mejora: data augmentation por traducción (ruso, tailandés, turco)

Del análisis de accuracy por idioma de arriba, los tres peores fueron ruso (80.5%), tailandés (81.2%) y turco (81.5%), todos por debajo de la accuracy global (87.7%). Probamos si dar al modelo más ejemplos de entrenamiento en esos idiomas ayuda.

Generamos ~300 ejemplos sintéticos por idioma traduciendo premise/hypothesis desde inglés (el idioma con más datos en el train set y la mejor calidad de traducción automática hacia estos tres) con [`facebook/nllb-200-distilled-600M`](https://huggingface.co/facebook/nllb-200-distilled-600M), un modelo many-to-many que cubre ruso, tailandés y turco directamente desde inglés en un único modelo. La label original se mantiene sin cambios: traducir premise e hypothesis no cambia la relación de entailment/neutral/contradiction entre ellas.

Los ~900 ejemplos aumentados se añaden **solo al split de entrenamiento** (nunca a `val_idx`), para que la comparación de accuracy antes/después siga midiéndose sobre el mismo conjunto de validación que ya usamos.

```python
from transformers import AutoModelForSeq2SeqLM

translator_name = 'facebook/nllb-200-distilled-600M'
translator_tokenizer = AutoTokenizer.from_pretrained(translator_name)

# liberamos temporalmente la VRAM de los modelos ya entrenados: no se vuelven a usar
model.to('cpu')
model_nli.to('cpu')
torch.cuda.empty_cache()

translator_model = AutoModelForSeq2SeqLM.from_pretrained(translator_name, dtype=torch.float16).to(device)
translator_model.eval()

# codigo NLLB de cada idioma objetivo (siempre traduciendo desde eng_Latn)
NLLB_LANG_CODES = {
    'Russian': ('ru', 'rus_Cyrl'),
    'Thai': ('th', 'tha_Thai'),
    'Turkish': ('tr', 'tur_Latn'),
}
N_AUG_PER_LANG = 300
translate_batch_size = 16

def translate_to(texts, tgt_code):
    translator_tokenizer.src_lang = 'eng_Latn'
    translations = []
    for i in range(0, len(texts), translate_batch_size):
        batch = texts[i:i + translate_batch_size]
        inputs = translator_tokenizer(
            batch, return_tensors='pt', padding=True, truncation=True, max_length=128
        ).to(device)
        with torch.no_grad():
            generated = translator_model.generate(
                **inputs,
                forced_bos_token_id=translator_tokenizer.convert_tokens_to_ids(tgt_code),
                max_new_tokens=128,
                num_beams=1)
        translations.extend(translator_tokenizer.batch_decode(generated, skip_special_tokens=True))
    return translations

english_train_pool = train.iloc[train_idx]
english_train_pool = english_train_pool[english_train_pool.language == 'English']

aug_rows = []
for lang_idx, (language, (lang_abv, tgt_code)) in enumerate(NLLB_LANG_CODES.items()):
    sample = english_train_pool.sample(n=N_AUG_PER_LANG, random_state=SEED + lang_idx)
    translated_premises = translate_to(list(sample.premise.values), tgt_code)
    translated_hypotheses = translate_to(list(sample.hypothesis.values), tgt_code)
    for j, (premise, hypothesis, label) in enumerate(zip(translated_premises, translated_hypotheses, sample.label.values)):
        aug_rows.append({
            'id': f'aug_{lang_abv}_{j}', 'premise': premise, 'hypothesis': hypothesis,
            'lang_abv': lang_abv, 'language': language, 'label': label,
        })
    print(f'{language}: {len(sample)} ejemplos traducidos desde ingles')

aug_df = pd.DataFrame(aug_rows)

# ya no necesitamos el traductor: liberamos su VRAM (model/model_nli se quedan en CPU, no hacen falta mas)
del translator_model
torch.cuda.empty_cache()

aug_df.sample(5, random_state=SEED)
```

Con los ~900 ejemplos aumentados construimos un nuevo set de entrenamiento: el train original (`train_idx`) más `aug_df`. El split de validación (`val_idx` / `val_loader_nli`) no se toca, así medimos el efecto real de la augmentación sobre el mismo conjunto de evaluación usado hasta ahora.

Entrenamos un modelo `model_nli_aug` nuevo desde cero (mismo checkpoint preentrenado `mDeBERTa-v3-base-mnli-xnli`), con la misma configuración que dio el mejor resultado hasta ahora — `K=3` capas superiores entrenables, `dropout=0.4`, mean pooling, `weight_decay=0.1` — para que la única diferencia frente al modelo anterior sea la augmentación de datos, y cualquier cambio en accuracy se le pueda atribuir a eso. Con más datos (~10.600 filas vs ~9.700) le damos algo más de margen de entrenamiento: `epochs=5` con `patience=3`, en vez de los 2 fijos que usamos antes para reproducir la mejor época ya conocida.

```python
train_aug_df = pd.concat([train.iloc[train_idx], aug_df], ignore_index=True)
train_labels_aug = torch.tensor(train_aug_df.label.values, dtype=torch.long)

train_input_nli_aug = bert_encode(train_aug_df.premise.values, train_aug_df.hypothesis.values, tokenizer_nli)
train_dataset_nli_aug = NLIDataset(train_input_nli_aug, train_labels_aug)
train_loader_nli_aug = DataLoader(train_dataset_nli_aug, batch_size=batch_size, shuffle=True)

print(f'Train original: {len(train_idx)} filas -> train aumentado: {len(train_aug_df)} filas (+{len(aug_df)})')
```

```python
model_nli_aug = NLIModel(model_name_nli, dropout=0.4, pooling='mean').float().to(device)

for name, param in model_nli_aug.bert.named_parameters():
    if name.startswith('encoder.layer.'):
        layer_idx = int(name.split('.')[2])
        param.requires_grad = layer_idx >= (model_nli_aug.bert.config.num_hidden_layers - K)
    else:
        param.requires_grad = False

for param in model_nli_aug.classifier.parameters():
    param.requires_grad = True

trainable_params_nli_aug = sum(p.numel() for p in model_nli_aug.parameters() if p.requires_grad)
print(f'Parametros entrenables: {trainable_params_nli_aug:,}')
```

```python
optimizer_nli_aug = torch.optim.AdamW(
    filter(lambda p: p.requires_grad, model_nli_aug.parameters()), lr=2e-5, weight_decay=0.1)

epochs_nli_aug = 5
patience_nli_aug = 3
total_steps_nli_aug = len(train_loader_nli_aug) * epochs_nli_aug
scheduler_nli_aug = get_linear_schedule_with_warmup(
    optimizer_nli_aug,
    num_warmup_steps=int(0.1 * total_steps_nli_aug),
    num_training_steps=total_steps_nli_aug)

best_val_loss_nli_aug = float('inf')
nli_aug_best_val_accuracy = None
epochs_no_improve_nli_aug = 0
best_epoch_nli_aug = None
saved_epochs_nli_aug = []  # que epochs quedaron guardadas en disco, para el checkpoint averaging de mas abajo

for epoch in range(epochs_nli_aug):
    model_nli_aug.train()
    train_loss, train_correct, train_total = 0.0, 0, 0
    for batch in train_loader_nli_aug:
        optimizer_nli_aug.zero_grad()
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        token_type_ids = batch['token_type_ids'].to(device)
        batch_labels = batch['label'].to(device)

        logits = model_nli_aug(input_ids, attention_mask, token_type_ids)
        loss = criterion(logits, batch_labels)
        loss.backward()
        torch.nn.utils.clip_grad_norm_(model_nli_aug.parameters(), max_norm=1.0)
        optimizer_nli_aug.step()
        scheduler_nli_aug.step()

        train_loss += loss.item() * batch_labels.size(0)
        train_correct += (logits.argmax(dim=-1) == batch_labels).sum().item()
        train_total += batch_labels.size(0)

    model_nli_aug.eval()
    val_loss, val_correct, val_total = 0.0, 0, 0
    with torch.no_grad():
        for batch in val_loader_nli:
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            token_type_ids = batch['token_type_ids'].to(device)
            batch_labels = batch['label'].to(device)

            logits = model_nli_aug(input_ids, attention_mask, token_type_ids)
            loss = criterion(logits, batch_labels)

            val_loss += loss.item() * batch_labels.size(0)
            val_correct += (logits.argmax(dim=-1) == batch_labels).sum().item()
            val_total += batch_labels.size(0)

    nli_aug_val_accuracy = val_correct / val_total
    avg_val_loss_nli_aug = val_loss / val_total
    print(f'Epoch {epoch+1}/{epochs_nli_aug} - '
          f'loss: {train_loss/train_total:.4f} - accuracy: {train_correct/train_total:.4f} - '
          f'val_loss: {avg_val_loss_nli_aug:.4f} - val_accuracy: {nli_aug_val_accuracy:.4f}')

    # Checkpointing: "last" se sobrescribe cada epoch, "best" solo cuando mejora la val_loss,
    # y ademas guardamos cada epoch por separado (nli_aug_epoch{N}.pt) para poder promediar pesos
    # de varias epochs cercanas a la mejor (checkpoint averaging / SWA-style), mas abajo.
    torch.save(model_nli_aug.state_dict(), os.path.join(checkpoint_dir, 'nli_aug_last.pt'))
    torch.save(model_nli_aug.state_dict(), os.path.join(checkpoint_dir, f'nli_aug_epoch{epoch+1}.pt'))
    saved_epochs_nli_aug.append(epoch + 1)

    if avg_val_loss_nli_aug < best_val_loss_nli_aug:
        best_val_loss_nli_aug = avg_val_loss_nli_aug
        nli_aug_best_val_accuracy = nli_aug_val_accuracy
        best_epoch_nli_aug = epoch + 1
        epochs_no_improve_nli_aug = 0
        torch.save(model_nli_aug.state_dict(), os.path.join(checkpoint_dir, 'nli_aug_best.pt'))
        print(f'  -> nuevo mejor val_loss ({best_val_loss_nli_aug:.4f}), guardado en nli_aug_best.pt')
    else:
        epochs_no_improve_nli_aug += 1
        print(f'  -> val_loss sin mejora desde hace {epochs_no_improve_nli_aug} epoch(s) (mejor: {best_val_loss_nli_aug:.4f})')
        if epochs_no_improve_nli_aug >= patience_nli_aug:
            print(f'Early stopping: sin mejora en val_loss durante {patience_nli_aug} epochs consecutivos.')
            break
```

Recargamos el mejor checkpoint (`nli_aug_best.pt`) y evaluamos sobre el mismo `val_loader_nli` de siempre, para comparar accuracy por idioma antes/después de la augmentación — sobre todo en ruso, tailandés y turco, que es lo que queríamos mejorar.

```python
model_nli_aug.load_state_dict(torch.load(os.path.join(checkpoint_dir, 'nli_aug_best.pt')))
model_nli_aug.eval()

val_preds_nli_aug, val_true_nli_aug = [], []
with torch.no_grad():
    for batch in val_loader_nli:
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        token_type_ids = batch['token_type_ids'].to(device)
        batch_labels = batch['label'].to(device)

        logits = model_nli_aug(input_ids, attention_mask, token_type_ids)
        val_preds_nli_aug.extend(logits.argmax(dim=-1).cpu().tolist())
        val_true_nli_aug.extend(batch_labels.cpu().tolist())

lang_df_aug = pd.DataFrame({'language': val_lang_nli, 'true': val_true_nli_aug, 'pred': val_preds_nli_aug})
lang_df_aug['correct'] = lang_df_aug['true'] == lang_df_aug['pred']
lang_acc_aug = lang_df_aug.groupby('language')['correct'].agg(accuracy='mean', n='size')

before_after = lang_acc[['accuracy']].rename(columns={'accuracy': 'antes'}).join(
    lang_acc_aug[['accuracy']].rename(columns={'accuracy': 'despues'}))
before_after['delta'] = before_after['despues'] - before_after['antes']
before_after.loc['GLOBAL'] = [overall_acc_nli, lang_df_aug['correct'].mean(),
                               lang_df_aug['correct'].mean() - overall_acc_nli]
before_after.sort_values('delta', ascending=False)
```

```python
plot_langs = ['Russian', 'Thai', 'Turkish', 'GLOBAL']
plot_df = before_after.loc[plot_langs]

x = np.arange(len(plot_langs))
width = 0.35

fig, ax = plt.subplots(figsize=(8, 5))
bars_before = ax.bar(x - width / 2, plot_df['antes'], width, label='Antes (sin augmentation)',
                      color='#2a78d6', zorder=3)
bars_after = ax.bar(x + width / 2, plot_df['despues'], width, label='Despues (+900 traducidos)',
                     color='#008300', zorder=3)

ax.set_xticks(x)
ax.set_xticklabels(plot_langs)
ax.set_ylabel('Accuracy (val, mDeBERTa best checkpoint)')
ax.set_title('Efecto de la data augmentation por traducción')
ax.set_ylim(0, 1)
ax.grid(axis='y', color='#e1e0d9', linewidth=0.8, zorder=0)
ax.set_axisbelow(True)
for spine in ['top', 'right']:
    ax.spines[spine].set_visible(False)
ax.legend(frameon=False, loc='upper center', bbox_to_anchor=(0.5, -0.12), ncol=2)

for bars in (bars_before, bars_after):
    for bar in bars:
        ax.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.01, f'{bar.get_height():.1%}',
                ha='center', fontsize=9, color='#0b0b0b')

plt.tight_layout()
plt.show()
```

**Resultado:** la augmentación ayudó, pero de forma desigual entre los tres idiomas objetivo — **ruso 80.5% → 85.4%** (+4.9 puntos, el idioma que más se benefició), **turco 81.5% → 83.3%** (+1.9 puntos), pero **tailandés se quedó plano en 81.2%** (sin cambio). La accuracy global subió ligeramente (87.7% → 88.4%, +0.7 puntos), con algún efecto colateral: chino bajó 4.1 puntos y unos pocos idiomas más (búlgaro, hindi, francés, urdu) perdieron 1-1.5 puntos — con n≈65-90 por idioma esto es del orden del ruido de muestreo que ya habíamos señalado, no necesariamente una regresión real.

Que tailandés no mejorara nada es la señal más interesante: puede ser que la traducción automática inglés→tailandés de NLLB sea de peor calidad que hacia ruso/turco (idiomas tipológicamente más cercanos al inglés), introduciendo ruido en vez de señal útil, o que 300 ejemplos no sean suficientes para ese idioma en concreto. Antes de aumentar el volumen de augmentación para tailandés, valdría la pena inspeccionar a mano una muestra de las traducciones generadas para descartar que sean de mala calidad.

## Ensemble: promediar `nli_best.pt` (sin augmentation) y `nli_aug_best.pt` (con augmentation)

Ya tenemos dos checkpoints entrenados sobre el mismo split de validación: el modelo original y el entrenado con los ~900 ejemplos traducidos. Antes de plantear cualquier reentrenamiento nuevo, probamos el ensemble más simple y barato posible — no hace falta entrenar nada más, solo recargar ambos checkpoints y promediar sus probabilidades softmax en cada ejemplo de `val_loader_nli` (`(softmax_a + softmax_b) / 2`, luego argmax).

La lógica: son dos modelos con la misma arquitectura pero pesos distintos (uno vio más ejemplos de ruso/tailandés/turco que el otro), así que sus errores no tienen por qué coincidir — si aciertan en ejemplos distintos, promediarlos puede superar a cualquiera de los dos por separado.

```python
model_nli.to(device)
model_nli.eval()
model_nli_aug.eval()

val_true_ens, val_preds_ens = [], []
with torch.no_grad():
    for batch in val_loader_nli:
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        token_type_ids = batch['token_type_ids'].to(device)
        batch_labels = batch['label'].to(device)

        probs_a = torch.softmax(model_nli(input_ids, attention_mask, token_type_ids), dim=-1)
        probs_b = torch.softmax(model_nli_aug(input_ids, attention_mask, token_type_ids), dim=-1)
        ensemble_probs = (probs_a + probs_b) / 2

        val_preds_ens.extend(ensemble_probs.argmax(dim=-1).cpu().tolist())
        val_true_ens.extend(batch_labels.cpu().tolist())

lang_df_ens = pd.DataFrame({'language': val_lang_nli, 'true': val_true_ens, 'pred': val_preds_ens})
lang_df_ens['correct'] = lang_df_ens['true'] == lang_df_ens['pred']
lang_acc_ens = lang_df_ens.groupby('language')['correct'].agg(accuracy='mean', n='size')
overall_acc_ens = lang_df_ens['correct'].mean()

comparison_3way = pd.DataFrame({
    'sin_augmentation': lang_acc['accuracy'],
    'con_augmentation': lang_acc_aug['accuracy'],
    'ensemble': lang_acc_ens['accuracy'],
})
comparison_3way.loc['GLOBAL'] = [overall_acc_nli, lang_df_aug['correct'].mean(), overall_acc_ens]
comparison_3way['mejor'] = comparison_3way[['sin_augmentation', 'con_augmentation', 'ensemble']].idxmax(axis=1)
comparison_3way.sort_values('ensemble', ascending=False)
```

```python
plot_langs = ['Russian', 'Thai', 'Turkish', 'GLOBAL']
plot_df = comparison_3way.loc[plot_langs]

x = np.arange(len(plot_langs))
width = 0.25

fig, ax = plt.subplots(figsize=(9, 5))
bars_1 = ax.bar(x - width, plot_df['sin_augmentation'], width, label='Sin augmentation',
                 color='#2a78d6', zorder=3)
bars_2 = ax.bar(x, plot_df['con_augmentation'], width, label='Con augmentation',
                 color='#008300', zorder=3)
bars_3 = ax.bar(x + width, plot_df['ensemble'], width, label='Ensemble (promedio de los dos)',
                 color='#e87ba4', zorder=3)

ax.set_xticks(x)
ax.set_xticklabels(plot_langs)
ax.set_ylabel('Accuracy (val, mDeBERTa)')
ax.set_title('Sin augmentation vs con augmentation vs ensemble')
ax.set_ylim(0, 1)
ax.grid(axis='y', color='#e1e0d9', linewidth=0.8, zorder=0)
ax.set_axisbelow(True)
for spine in ['top', 'right']:
    ax.spines[spine].set_visible(False)
ax.legend(frameon=False, loc='upper center', bbox_to_anchor=(0.5, -0.12), ncol=3)

for bars in (bars_1, bars_2, bars_3):
    for bar in bars:
        ax.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.01, f'{bar.get_height():.1%}',
                ha='center', fontsize=8, color='#0b0b0b')

plt.tight_layout()
plt.show()
```

**Resultado: el ensemble no gana.** Global, `con_augmentation` sola (88.4%) sigue siendo mejor que el ensemble (88.1%), que a su vez supera a `sin_augmentation` sola (87.7%). Para ruso y turco el ensemble queda *entre* los dos modelos individuales, nunca por encima del mejor de los dos — al promediar, `model_nli` (que seguía siendo bastante peor en ruso, 80.5%) arrastra hacia abajo el 85.4% que había conseguido `model_nli_aug` en solitario (ruso: ensemble 82.9%, un punto y medio por debajo de usar solo el modelo con augmentation).

La única excepción es tailandés: ahí `sin_augmentation` y `con_augmentation` empataban exactos en 81.2% (la augmentación no había movido ni un ejemplo), pero el ensemble sube a 82.6% — señal de que en ese idioma concreto los dos modelos sí aciertan en ejemplos distintos y se complementan un poco, aunque no compensa la pérdida en el resto de idiomas.

**Conclusión práctica:** para este par de checkpoints concretos, quedarse con `nli_aug_best.pt` (con augmentation) a secas es mejor que promediarlo con el modelo sin augmentation — tiene sentido, porque los dos modelos comparten casi todo el training set (solo difieren en ~900 de ~10.600 filas) y por tanto sus errores están muy correlacionados, justo lo contrario de lo que necesita un ensemble para sumar. Si se quiere seguir explorando ensembling, probablemente haga falta diversidad real entre los modelos (distintas seeds, distinto `K`, distinto pooling, o incluso arquitecturas distintas) en vez de dos variantes casi idénticas del mismo modelo.

## Mejora: domain-adaptive pretraining (MLM continuado) en tailandés

La augmentación por traducción (sección anterior) no movió la accuracy en tailandés ni un punto (81.2% → 81.2%), y ya descartamos que fuera un problema de `max_len` (0% de truncamiento en tailandés con el tokenizer de mDeBERTa). La hipótesis que quedó pendiente: la traducción automática inglés→tailandés de NLLB-600M puede ser de calidad más baja que hacia ruso/turco (idiomas más cercanos tipológicamente al inglés), así que los ~300 ejemplos traducidos podrían estar aportando ruido en vez de señal útil.

En vez de seguir por la vía de "más traducción", probamos una estrategia complementaria y no supervisada: continuar el pre-entrenamiento con masked language modeling (MLM) del backbone de mDeBERTa sobre texto tailandés **real** (no traducido), antes del fine-tuning supervisado de NLI. La idea es que el modelo ajuste sus representaciones internas al tailandés real sin depender de traducciones automáticas de calidad incierta — no hace falta ninguna etiqueta, solo texto en el idioma, así que el corpus puede incluir texto de `test.csv` (que no tiene labels) sin ningún problema.

Deliberadamente **no** incluimos aquí el texto tailandés traducido por NLLB (`aug_df`): si la hipótesis es que esas traducciones son de baja calidad, meterlas en el corpus de MLM reintroduciría el mismo ruido que estamos intentando evitar. Y por la misma razón de siempre, tampoco tocamos `val_idx` — ni con etiquetas ni sin ellas — para que la comparación de accuracy después siga siendo limpia.

Reutilizamos el mismo patrón de freezing que en el resto del notebook (`K=3` capas superiores del encoder entrenables) para que el MLM continuado adapte exactamente las mismas capas que luego se van a fine-tunear con las etiquetas de NLI.

```python
thai_train_mask = train.iloc[train_idx].language == 'Thai'
thai_train_texts = pd.concat([
    train.iloc[train_idx][thai_train_mask].premise,
    train.iloc[train_idx][thai_train_mask].hypothesis,
])

thai_test_mask = test.language == 'Thai'
thai_test_texts = pd.concat([
    test[thai_test_mask].premise,
    test[thai_test_mask].hypothesis,
])

dapt_texts = pd.concat([thai_train_texts, thai_test_texts]).dropna()
dapt_texts = dapt_texts[dapt_texts.str.strip() != ''].drop_duplicates().tolist()

print(f'Corpus tailandes para MLM continuado: {len(dapt_texts)} frases unicas '
      f'(train_idx: {thai_train_mask.sum()} filas, test.csv: {thai_test_mask.sum()} filas)')

from transformers import AutoModelForMaskedLM, DataCollatorForLanguageModeling

# sin padding fijo a max_len=128 aqui: las frases sueltas en tailandes son bastante mas cortas que los
# pares premise+hypothesis del resto del notebook, y el collator ya paddea dinamicamente por batch
# (importa para memoria: los logits de MLM son (batch, seq_len, vocab_size), asi que un seq_len mas
# corto reduce mucho el pico de VRAM en el backward)
dapt_encodings = tokenizer_nli(dapt_texts, truncation=True, max_length=max_len)

class MLMTextDataset(Dataset):
    def __init__(self, encodings):
        self.input_ids = encodings['input_ids']
        self.attention_mask = encodings['attention_mask']

    def __len__(self):
        return len(self.input_ids)

    def __getitem__(self, idx):
        return {'input_ids': self.input_ids[idx], 'attention_mask': self.attention_mask[idx]}

dapt_dataset = MLMTextDataset(dapt_encodings)
mlm_collator = DataCollatorForLanguageModeling(tokenizer=tokenizer_nli, mlm=True, mlm_probability=0.15)
dapt_batch_size = 8  # mas pequeno que el batch_size=32 general: el corpus es chico y asi controlamos el pico de VRAM
dapt_loader = DataLoader(dapt_dataset, batch_size=dapt_batch_size, shuffle=True, collate_fn=mlm_collator)
```

`AutoModelForMaskedLM` sobre este checkpoint concreto avisa de que faltan los pesos de `cls.predictions.*` (la cabeza de MLM) — es esperable: `mDeBERTa-v3-base-mnli-xnli` se guardó como modelo de clasificación (fine-tuneado para MNLI/XNLI), no con su cabeza de MLM original, así que Transformers la inicializa nueva. Es la misma situación que ya documentamos para el `classifier` de `NLIModel` (nota de fp16 vs fp32 en el CLAUDE.md): forzamos `.float()` al instanciar.

Igual que en el resto del notebook, congelamos embeddings + capas inferiores y dejamos entrenables solo las últimas `K=3` capas del encoder — aquí además la cabeza de MLM (`cls.*`), que es completamente nueva y tiene que entrenarse sí o sí.

```python
# los modelos ya entrenados (baseline, model_nli, model_nli_aug) no se vuelven a usar en el resto del
# notebook: los borramos del todo (no solo .to('cpu')) para liberar tambien la memoria que retenian sus
# optimizers/schedulers, que si no seguirian referenciando los parametros y evitando que se liberasen
import gc

for _name in ['model', 'model_nli', 'model_nli_aug',
              'optimizer', 'optimizer_nli', 'optimizer_nli_aug',
              'scheduler_nli', 'scheduler_nli_aug']:
    if _name in globals():
        del globals()[_name]
gc.collect()
torch.cuda.empty_cache()

model_dapt_backbone = AutoModelForMaskedLM.from_pretrained(model_name_nli).float().to(device)

for name, param in model_dapt_backbone.named_parameters():
    if '.encoder.layer.' in name:
        layer_idx = int(name.split('.encoder.layer.')[1].split('.')[0])
        param.requires_grad = layer_idx >= (model_dapt_backbone.config.num_hidden_layers - K)
    elif name.startswith('cls.'):
        param.requires_grad = True  # cabeza de MLM, nueva, siempre entrenable
    else:
        param.requires_grad = False

trainable_dapt = sum(p.numel() for p in model_dapt_backbone.parameters() if p.requires_grad)
total_dapt = sum(p.numel() for p in model_dapt_backbone.parameters())
print(f'Parametros entrenables (MLM continuado): {trainable_dapt:,} de {total_dapt:,} ({trainable_dapt/total_dapt:.1%})')
```

Corpus pequeño (unos cientos de frases), así que un entrenamiento corto: `epochs_dapt=3` con LR bajo (`5e-5`) para adaptar sin sobreescribir de golpe lo que el modelo ya sabe de NLI. No hay split de validación aquí — es una adaptación de dominio intermedia, no el paso donde decidimos qué pesos quedarse; el checkpoint que de verdad importa es el del fine-tuning supervisado posterior, que sí tiene su propio val_loss/early stopping como todo lo demás en el notebook.

```python
optimizer_dapt = torch.optim.AdamW(
    filter(lambda p: p.requires_grad, model_dapt_backbone.parameters()), lr=5e-5)

epochs_dapt = 3
model_dapt_backbone.train()
for epoch in range(epochs_dapt):
    epoch_loss, n_batches = 0.0, 0
    for batch in dapt_loader:
        optimizer_dapt.zero_grad()
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        mlm_labels = batch['labels'].to(device)

        outputs = model_dapt_backbone(input_ids=input_ids, attention_mask=attention_mask, labels=mlm_labels)
        loss = outputs.loss
        loss.backward()
        torch.nn.utils.clip_grad_norm_(model_dapt_backbone.parameters(), max_norm=1.0)
        optimizer_dapt.step()

        epoch_loss += loss.item()
        n_batches += 1

    print(f'DAPT epoch {epoch+1}/{epochs_dapt} - mlm_loss: {epoch_loss/n_batches:.4f}')
```

Con el backbone ya adaptado, construimos un `NLIModel` nuevo copiando esos pesos (en vez de partir otra vez del checkpoint original de Hugging Face) y lo fine-tuneamos igual que `model_nli` — mismo `K=3`, `dropout=0.4`, mean pooling, `weight_decay=0.1` — sobre `train_loader_nli` **sin** los ejemplos traducidos (`aug_df`), para aislar el efecto del DAPT del efecto de la augmentación por traducción, que ya sabemos que no ayudó en tailandés. Igual que con `model_nli_aug`, damos algo más de margen (`epochs=5`, `patience=3`) porque es un modelo nuevo desde este punto de partida, no una reproducción de una época ya conocida.

```python
model_nli_dapt = NLIModel(model_name_nli, dropout=0.4, pooling='mean').float().to(device)
model_nli_dapt.bert.load_state_dict(model_dapt_backbone.deberta.state_dict())

for name, param in model_nli_dapt.bert.named_parameters():
    if name.startswith('encoder.layer.'):
        layer_idx = int(name.split('.')[2])
        param.requires_grad = layer_idx >= (model_nli_dapt.bert.config.num_hidden_layers - K)
    else:
        param.requires_grad = False

for param in model_nli_dapt.classifier.parameters():
    param.requires_grad = True

# ya copiamos los pesos adaptados a model_nli_dapt.bert; no hace falta conservar el backbone de MLM
del model_dapt_backbone
torch.cuda.empty_cache()

trainable_dapt_ft = sum(p.numel() for p in model_nli_dapt.parameters() if p.requires_grad)
print(f'Parametros entrenables (fine-tuning tras DAPT): {trainable_dapt_ft:,}')
```

```python
optimizer_nli_dapt = torch.optim.AdamW(
    filter(lambda p: p.requires_grad, model_nli_dapt.parameters()), lr=2e-5, weight_decay=0.1)

epochs_nli_dapt = 5
patience_nli_dapt = 3
total_steps_nli_dapt = len(train_loader_nli) * epochs_nli_dapt
scheduler_nli_dapt = get_linear_schedule_with_warmup(
    optimizer_nli_dapt,
    num_warmup_steps=int(0.1 * total_steps_nli_dapt),
    num_training_steps=total_steps_nli_dapt)

best_val_loss_nli_dapt = float('inf')
nli_dapt_best_val_accuracy = None
epochs_no_improve_nli_dapt = 0

for epoch in range(epochs_nli_dapt):
    model_nli_dapt.train()
    train_loss, train_correct, train_total = 0.0, 0, 0
    for batch in train_loader_nli:
        optimizer_nli_dapt.zero_grad()
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        token_type_ids = batch['token_type_ids'].to(device)
        batch_labels = batch['label'].to(device)

        logits = model_nli_dapt(input_ids, attention_mask, token_type_ids)
        loss = criterion(logits, batch_labels)
        loss.backward()
        torch.nn.utils.clip_grad_norm_(model_nli_dapt.parameters(), max_norm=1.0)
        optimizer_nli_dapt.step()
        scheduler_nli_dapt.step()

        train_loss += loss.item() * batch_labels.size(0)
        train_correct += (logits.argmax(dim=-1) == batch_labels).sum().item()
        train_total += batch_labels.size(0)

    model_nli_dapt.eval()
    val_loss, val_correct, val_total = 0.0, 0, 0
    with torch.no_grad():
        for batch in val_loader_nli:
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            token_type_ids = batch['token_type_ids'].to(device)
            batch_labels = batch['label'].to(device)

            logits = model_nli_dapt(input_ids, attention_mask, token_type_ids)
            loss = criterion(logits, batch_labels)

            val_loss += loss.item() * batch_labels.size(0)
            val_correct += (logits.argmax(dim=-1) == batch_labels).sum().item()
            val_total += batch_labels.size(0)

    nli_dapt_val_accuracy = val_correct / val_total
    avg_val_loss_nli_dapt = val_loss / val_total
    print(f'Epoch {epoch+1}/{epochs_nli_dapt} - '
          f'loss: {train_loss/train_total:.4f} - accuracy: {train_correct/train_total:.4f} - '
          f'val_loss: {avg_val_loss_nli_dapt:.4f} - val_accuracy: {nli_dapt_val_accuracy:.4f}')

    # Checkpointing: "last" se sobrescribe cada epoch, "best" solo cuando mejora la val_loss
    torch.save(model_nli_dapt.state_dict(), os.path.join(checkpoint_dir, 'nli_dapt_last.pt'))

    if avg_val_loss_nli_dapt < best_val_loss_nli_dapt:
        best_val_loss_nli_dapt = avg_val_loss_nli_dapt
        nli_dapt_best_val_accuracy = nli_dapt_val_accuracy
        epochs_no_improve_nli_dapt = 0
        torch.save(model_nli_dapt.state_dict(), os.path.join(checkpoint_dir, 'nli_dapt_best.pt'))
        print(f'  -> nuevo mejor val_loss ({best_val_loss_nli_dapt:.4f}), guardado en nli_dapt_best.pt')
    else:
        epochs_no_improve_nli_dapt += 1
        print(f'  -> val_loss sin mejora desde hace {epochs_no_improve_nli_dapt} epoch(s) (mejor: {best_val_loss_nli_dapt:.4f})')
        if epochs_no_improve_nli_dapt >= patience_nli_dapt:
            print(f'Early stopping: sin mejora en val_loss durante {patience_nli_dapt} epochs consecutivos.')
            break
```

Recargamos el mejor checkpoint (`nli_dapt_best.pt`) y evaluamos sobre el mismo `val_loader_nli` de siempre, comparando accuracy por idioma frente a `sin_augmentation` y `con_augmentation` (traducción) — el idioma de interés es tailandés, donde la traducción no había movido nada.

```python
model_nli_dapt.load_state_dict(torch.load(os.path.join(checkpoint_dir, 'nli_dapt_best.pt')))
model_nli_dapt.eval()

val_preds_nli_dapt, val_true_nli_dapt = [], []
with torch.no_grad():
    for batch in val_loader_nli:
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        token_type_ids = batch['token_type_ids'].to(device)
        batch_labels = batch['label'].to(device)

        logits = model_nli_dapt(input_ids, attention_mask, token_type_ids)
        val_preds_nli_dapt.extend(logits.argmax(dim=-1).cpu().tolist())
        val_true_nli_dapt.extend(batch_labels.cpu().tolist())

lang_df_dapt = pd.DataFrame({'language': val_lang_nli, 'true': val_true_nli_dapt, 'pred': val_preds_nli_dapt})
lang_df_dapt['correct'] = lang_df_dapt['true'] == lang_df_dapt['pred']
lang_acc_dapt = lang_df_dapt.groupby('language')['correct'].agg(accuracy='mean', n='size')
overall_acc_dapt = lang_df_dapt['correct'].mean()

comparison_dapt = pd.DataFrame({
    'sin_augmentation': lang_acc['accuracy'],
    'con_augmentation_traduccion': lang_acc_aug['accuracy'],
    'con_dapt_tailandes': lang_acc_dapt['accuracy'],
})
comparison_dapt.loc['GLOBAL'] = [overall_acc_nli, lang_df_aug['correct'].mean(), overall_acc_dapt]
comparison_dapt.sort_values('con_dapt_tailandes', ascending=False)
```

```python
plot_langs = ['Thai', 'Russian', 'Turkish', 'GLOBAL']
plot_df = comparison_dapt.loc[plot_langs]

x = np.arange(len(plot_langs))
width = 0.25

fig, ax = plt.subplots(figsize=(9, 5))
bars_1 = ax.bar(x - width, plot_df['sin_augmentation'], width, label='Sin augmentation',
                 color='#2a78d6', zorder=3)
bars_2 = ax.bar(x, plot_df['con_augmentation_traduccion'], width, label='Con augmentation (traduccion)',
                 color='#008300', zorder=3)
bars_3 = ax.bar(x + width, plot_df['con_dapt_tailandes'], width, label='Con DAPT (MLM tailandes)',
                 color='#e8a33d', zorder=3)

ax.set_xticks(x)
ax.set_xticklabels(plot_langs)
ax.set_ylabel('Accuracy (val, mDeBERTa)')
ax.set_title('Traduccion vs domain-adaptive pretraining en tailandes')
ax.set_ylim(0, 1)
ax.grid(axis='y', color='#e1e0d9', linewidth=0.8, zorder=0)
ax.set_axisbelow(True)
for spine in ['top', 'right']:
    ax.spines[spine].set_visible(False)
ax.legend(frameon=False, loc='upper center', bbox_to_anchor=(0.5, -0.12), ncol=3)

for bars in (bars_1, bars_2, bars_3):
    for bar in bars:
        ax.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.01, f'{bar.get_height():.1%}',
                ha='center', fontsize=8, color='#0b0b0b')

plt.tight_layout()
plt.show()
```

**Resultado: el DAPT tampoco mueve tailandés.** La `mlm_loss` bajó de forma consistente durante el pre-entrenamiento continuado (11.32 → 9.17 → 8.87 en 3 épocas), así que el modelo sí estaba aprendiendo algo del corpus tailandés. Pero en el fine-tuning supervisado posterior, tailandés se queda en **81.16% — exactamente la misma cifra** que sin augmentation y que con augmentation por traducción (56/69 aciertos en los tres casos). Dos estrategias completamente distintas (traducción NLLB y MLM no supervisado) convergen al mismo resultado nulo en el mismo idioma, lo cual es información real: apunta a que el cuello de botella en tailandés no es de representación léxica/de dominio (lo que ambas estrategias atacan), sino en otro sitio — quizás la propia dificultad de los ejemplos tailandeses (ambigüedad genuina premise/hypothesis) o algo específico del classifier head para ese idioma que ni más datos ni mejor representación del encoder consiguen arreglar.

Efecto colateral a anotar: ruso (que el DAPT ni tocó, solo entrenamos con datos tailandeses no supervisados) bajó de 80.5% a 79.3% frente a `sin_augmentation` — con n=82 esto es del orden del ruido de muestreo habitual entre reentrenamientos con distinta inicialización del classifier y distinto shuffling, no necesariamente un efecto causal del DAPT. La accuracy global con DAPT (87.83%) queda ligeramente por encima de `sin_augmentation` (87.71%, +0.12pts) pero muy por debajo de lo que consiguió la augmentación por traducción (88.41%, +0.70pts) — para el idioma que sí respondía a intervención (ruso), traducir ejemplos reales sigue siendo mejor palanca que adaptar el dominio de forma no supervisada.

**Conclusión práctica:** con un corpus de partida tan pequeño (913 frases), el DAPT no fue la solución para tailandés. El siguiente paso razonable no es escalar el DAPT (más épocas o más texto tailandés no etiquetado, que tampoco abunda en este dataset), sino la inspección manual que ya habíamos apuntado como pendiente: mirar a mano una muestra de los errores del modelo en tailandés — si son ambigüedades genuinas del propio ejemplo (premise/hypothesis donde razonablemente un humano dudaría) en vez de errores por mala representación del idioma, ninguna de las dos estrategias de este notebook iba a poder arreglarlo.

## Verificación: fuga de datos (leakage) entre train y validación

La guía de DeBERTa señala la fuga de datos como una causa típica de "resultados inestables" y pide verificarla explícitamente en vez de asumir que un split aleatorio es automáticamente limpio. No lo habíamos comprobado hasta ahora — todas las cifras de este notebook (baseline, mDeBERTa, augmentation, ensemble, DAPT) usan el mismo `train_idx`/`val_idx` de `train_test_split(random_state=42)`, así que si hay fuga, contamina *todas* las comparaciones por igual, no solo una.

Dos chequeos distintos:
1. **Duplicados exactos** — misma `(premise, hypothesis)` en ambos splits. Sería fuga directa de un ejemplo etiquetado completo.
2. **Premises compartidas** — el dataset de esta competición genera varias filas (hasta 3: entailment/neutral/contradiction) a partir del mismo `premise`, y `train_test_split` reparte filas al azar sin agrupar por `premise`. Si una fila de un `premise` cae en train y otra fila del mismo `premise` cae en val, el modelo ya ha visto ese `premise` exacto durante el entrenamiento (con otra `hypothesis`/label) antes de evaluarse sobre él — una fuga más sutil pero real, típica de datasets de NLI construidos por plantillas.

Para medir el impacto sin reentrenar nada, reutilizamos `val_true_nli`/`val_preds_nli` (el mejor checkpoint de `model_nli`, sin augmentation) y separamos el split de validación en dos grupos: filas cuyo `premise` también aparece en `train_idx` ("leaky") y filas cuyo `premise` es genuinamente nuevo ("limpio"). Si la accuracy es notablemente más alta en el grupo leaky, la cifra global de accuracy reportada en todo el notebook estaría inflada por esta fuga.

```python
train_pairs = set(zip(train.iloc[train_idx].premise, train.iloc[train_idx].hypothesis))
val_pairs = set(zip(train.iloc[val_idx].premise, train.iloc[val_idx].hypothesis))
exact_dupes = train_pairs & val_pairs

train_premises_set = set(train.iloc[train_idx].premise)
val_premises_set = set(train.iloc[val_idx].premise)
premise_overlap = train_premises_set & val_premises_set

print(f'Duplicados exactos (mismo premise+hypothesis) entre train_idx y val_idx: {len(exact_dupes)}')
print(f'Premises de val_idx que tambien aparecen en train_idx (con otra hypothesis/label): '
      f'{len(premise_overlap)} de {len(val_premises_set)} ({len(premise_overlap)/len(val_premises_set):.1%})')
```

```python
val_premises_arr = train.iloc[val_idx].premise.values
is_leaky = np.isin(val_premises_arr, list(train_premises_set))

leak_df = pd.DataFrame({
    'language': val_lang_nli,
    'true': val_true_nli,
    'pred': val_preds_nli,
    'leaky': is_leaky,
})
leak_df['correct'] = leak_df['true'] == leak_df['pred']

acc_by_leak = leak_df.groupby('leaky')['correct'].agg(accuracy='mean', n='size')
acc_by_leak.index = acc_by_leak.index.map({True: 'leaky (premise vista en train)', False: 'limpio (premise no vista)'})
print(acc_by_leak)
print()

leaky_ratio_by_lang = leak_df.groupby('language')['leaky'].mean().sort_values(ascending=False)
print('Proporcion de filas "leaky" por idioma (val_idx):')
print(leaky_ratio_by_lang)
```

```python
colors = ['#2a78d6' if 'limpio' in lbl else '#d03b3b' for lbl in acc_by_leak.index]

fig, ax = plt.subplots(figsize=(6, 5))
bars = ax.bar(acc_by_leak.index, acc_by_leak['accuracy'], color=colors, zorder=3)
ax.set_ylabel('Accuracy (val, mDeBERTa sin augmentation)')
ax.set_title('Accuracy segun si la premise tambien aparece en train')
ax.set_ylim(0, 1)
ax.grid(axis='y', color='#e1e0d9', linewidth=0.8, zorder=0)
ax.set_axisbelow(True)
for spine in ['top', 'right']:
    ax.spines[spine].set_visible(False)
for bar, acc, n in zip(bars, acc_by_leak['accuracy'], acc_by_leak['n']):
    ax.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.01, f'{acc:.1%} (n={n})',
            ha='center', fontsize=9, color='#0b0b0b')
plt.tight_layout()
plt.show()
```

**Resultado: no hay fuga directa, y la fuga de premises compartidas no infla la accuracy.** Cero duplicados exactos `(premise, hypothesis)` entre `train_idx` y `val_idx` — el split no repite ningún ejemplo etiquetado completo. Sí hay overlap de `premise` en el sentido más suave: el 47.3% de las premises de validación (1054 de 2226) también aparecen en `train_idx` con otra `hypothesis`/label, resultado esperable dado que el dataset genera hasta 3 filas por `premise` y el split no agrupa por ese campo.

Lo importante es que esta fuga **no se traduce en accuracy inflada**: el subconjunto "leaky" (premise vista en train) obtiene 87.54% (n=1164) frente a 87.86% (n=1260) del subconjunto genuinamente "limpio" — prácticamente idéntico, y si acaso *ligeramente peor* en el grupo leaky, no mejor. Es decir, haber visto la misma premise con otra hypothesis durante el entrenamiento no le da al modelo ninguna ventaja medible al evaluar una hypothesis nueva sobre esa premise — tiene sentido, porque la tarea de NLI depende de la relación premise-hypothesis, no de memorizar el contenido de la premise por sí sola.

Dato adicional: el overlap está muy concentrado en inglés (77.6% de sus filas de validación son "leaky", frente a solo 10.1% en tailandés y 3.7% en turco) — inglés es, con diferencia, el idioma con más filas en el dataset, así que es el que más premises repite entre splits. Para los idiomas que nos interesaban en las secciones de augmentation y DAPT (ruso, tailandés, turco), la proporción de leakage es baja.

**Conclusión práctica:** las cifras de accuracy reportadas en todo el notebook (87-88% global, y las comparaciones por idioma) no están infladas por esta forma de fuga — no hace falta rehacer el split ni desconfiar de los resultados ya obtenidos. Dicho esto, si en el futuro se reentrena con un dataset de NLI distinto donde el groupeo por `premise` sea más pronunciado, vale la pena repetir este chequeo antes de confiar en el número: aquí salió limpio, pero no es algo que se pueda dar por hecho solo porque `train_test_split` sea aleatorio.

## Mejora: checkpoint averaging (SWA-style) sobre `model_nli_aug`

Ni la augmentación por traducción en tailandés ni el domain-adaptive pretraining movieron esa cifra — dos intentos independientes convergiendo al mismo resultado nulo es buena señal de que seguir invirtiendo en casos concretos (tailandés en particular) no es el camino más eficiente ahora mismo. Cambiamos de estrategia: en vez de perseguir un idioma, probamos una técnica que actúa sobre el modelo en conjunto — checkpoint averaging.

A diferencia del ensemble que ya probamos (que promediaba dos modelos entrenados con datos distintos y no ganó, porque sus errores estaban demasiado correlacionados), aquí promediamos los **pesos** de varias épocas del **mismo** entrenamiento — concretamente las de `model_nli_aug`, el mejor modelo global hasta ahora (88.4%). La idea es distinta: en vez de buscar diversidad entre modelos, aprovechamos que los pesos de épocas consecutivas cercanas al mínimo de val_loss suelen estar en una región razonablemente plana del espacio de pérdida, y promediarlos puede dar una solución algo más robusta que cualquiera de las épocas individuales — sin entrenar nada nuevo, solo reutilizando checkpoints que ya guardamos época a época durante el entrenamiento (`nli_aug_epoch{N}.pt`, añadido específicamente para esto).

Ventana de promediado: la mejor época ± 1, limitada a las que realmente se guardaron en disco (si la mejor fue la primera o la última época entrenada, la ventana se recorta en vez de salirse de rango).

```python
# model_nli_dapt ya no se usa (la seccion de DAPT y la de leakage terminaron); lo liberamos antes
# de cargar varios checkpoints de ~1.1GB cada uno para el promediado, para no arriesgar VRAM
import gc
if 'model_nli_dapt' in globals():
    del model_nli_dapt
gc.collect()
torch.cuda.empty_cache()

avg_window = [e for e in [best_epoch_nli_aug - 1, best_epoch_nli_aug, best_epoch_nli_aug + 1]
              if e in saved_epochs_nli_aug]
print(f'Promediando pesos de las epochs {avg_window} (mejor epoch: {best_epoch_nli_aug})')

# map_location='cpu': promediamos en CPU en vez de cargar varios checkpoints de golpe en la GPU
state_dicts = [torch.load(os.path.join(checkpoint_dir, f'nli_aug_epoch{e}.pt'), map_location='cpu')
               for e in avg_window]
averaged_state_dict = {
    key: sum(sd[key].float() for sd in state_dicts) / len(state_dicts)
    for key in state_dicts[0]
}
del state_dicts
gc.collect()

model_nli_swa = NLIModel(model_name_nli, dropout=0.4, pooling='mean').float().to(device)
model_nli_swa.load_state_dict(averaged_state_dict)
model_nli_swa.eval()
del averaged_state_dict

val_preds_nli_swa, val_true_nli_swa = [], []
with torch.no_grad():
    for batch in val_loader_nli:
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        token_type_ids = batch['token_type_ids'].to(device)
        batch_labels = batch['label'].to(device)

        logits = model_nli_swa(input_ids, attention_mask, token_type_ids)
        val_preds_nli_swa.extend(logits.argmax(dim=-1).cpu().tolist())
        val_true_nli_swa.extend(batch_labels.cpu().tolist())

lang_df_swa = pd.DataFrame({'language': val_lang_nli, 'true': val_true_nli_swa, 'pred': val_preds_nli_swa})
lang_df_swa['correct'] = lang_df_swa['true'] == lang_df_swa['pred']
lang_acc_swa = lang_df_swa.groupby('language')['correct'].agg(accuracy='mean', n='size')
overall_acc_swa = lang_df_swa['correct'].mean()

comparison_swa = pd.DataFrame({
    'con_augmentation (best checkpoint)': lang_acc_aug['accuracy'],
    'con_augmentation + checkpoint_averaging': lang_acc_swa['accuracy'],
})
comparison_swa.loc['GLOBAL'] = [lang_df_aug['correct'].mean(), overall_acc_swa]
comparison_swa['delta'] = (comparison_swa['con_augmentation + checkpoint_averaging']
                            - comparison_swa['con_augmentation (best checkpoint)'])
comparison_swa.sort_values('delta', ascending=False)
```

```python
plot_langs = ['Russian', 'Thai', 'Turkish', 'GLOBAL']
plot_df = comparison_swa.loc[plot_langs]

x = np.arange(len(plot_langs))
width = 0.35

fig, ax = plt.subplots(figsize=(8, 5))
bars_1 = ax.bar(x - width / 2, plot_df['con_augmentation (best checkpoint)'], width,
                 label='Con augmentation (best checkpoint)', color='#008300', zorder=3)
bars_2 = ax.bar(x + width / 2, plot_df['con_augmentation + checkpoint_averaging'], width,
                 label='+ checkpoint averaging', color='#7a4fd6', zorder=3)

ax.set_xticks(x)
ax.set_xticklabels(plot_langs)
ax.set_ylabel('Accuracy (val, mDeBERTa)')
ax.set_title('Checkpoint averaging sobre model_nli_aug')
ax.set_ylim(0, 1)
ax.grid(axis='y', color='#e1e0d9', linewidth=0.8, zorder=0)
ax.set_axisbelow(True)
for spine in ['top', 'right']:
    ax.spines[spine].set_visible(False)
ax.legend(frameon=False, loc='upper center', bbox_to_anchor=(0.5, -0.12), ncol=2)

for bars in (bars_1, bars_2):
    for bar in bars:
        ax.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.01, f'{bar.get_height():.1%}',
                ha='center', fontsize=9, color='#0b0b0b')

plt.tight_layout()
plt.show()
```

**Resultado: prácticamente un empate, ligeramente negativo en global.** Promediar los pesos de las épocas 2, 3 y 4 (la mejor fue la 3) da 88.37% global frente al 88.41% del mejor checkpoint solo — una diferencia de -0.04pts, dentro del ruido. Por idioma sí hay movimiento en ambas direcciones: turco (+1.9pts), búlgaro (+1.5), chino (+1.4) y francés (+1.3) mejoran, pero urdu (-1.2) e inglés (-0.3) empeoran, compensándose casi exactamente. Tailandés y ruso no se mueven nada.

Tiene sentido con el propio entrenamiento: las épocas 2, 3 y 4 ya tienen val_loss muy parecida entre sí (0.3497 / 0.3394 / 0.3497) — el LR decae con el schedule y el modelo ya está bastante convergido en esa zona, así que no hay suficiente diversidad entre esos tres checkpoints como para que promediarlos aporte algo. El truco de SWA suele funcionar mejor con LR constante o cíclico, donde los checkpoints exploran regiones más distintas del espacio de pesos — no es el caso aquí.

**Conclusión práctica:** para este entrenamiento concreto, quedarse con el mejor checkpoint (`nli_aug_best.pt`) es tan bueno como promediar, y más simple.

## Mejora: descongelamiento gradual del encoder (gradual unfreezing)

Hasta ahora todas las secciones de mDeBERTa fijan `K=3` capas superiores entrenables desde la época 1. Aquí probamos una rampa: la capacidad entrenable crece progresivamente en vez de estar fija desde el principio — época 1 solo entrena el classifier head nuevo (`K=0`), y cada época añade más capas del encoder.

**Primer intento (K: 0→3→6→9→12, hasta el encoder completo) provocó un reset del driver de NVIDIA** (evento `nvlddmkm` ID 153 en el Visor de sucesos de Windows — un TDR, Timeout Detection and Recovery: Windows resetea la GPU si un kernel CUDA individual tarda más del umbral por defecto de 2 segundos). El backward completo por las 12 capas de un modelo de disentangled attention como DeBERTa-v3 es lo bastante pesado como para dispararlo en este equipo. En vez de tocar el registro de Windows (`TdrDelay`, que requiere admin y reiniciar), recortamos la rampa a `K: 0→2→4→6→8` — sigue yendo bastante más allá del `K=3` que usa el resto del notebook, sin acercarse al límite que causó el reset.

Para que esto no salga más caro que el entrenamiento estático: se reutiliza el mismo presupuesto de épocas (`epochs=5`, igual que `model_nli_aug`) en vez de añadir fases extra, y el optimizer se construye **una sola vez** sobre todos los parámetros del encoder + classifier (no solo los descongelados en cada momento) — togglear `requires_grad` por época es gratis, y AdamW se salta automáticamente cualquier parámetro cuyo `.grad` sea `None`, así que no hace falta reconstruir el optimizer ni perder el momentum acumulado de las capas ya descongeladas. El forward pass ya recorre las 12 capas del encoder de todas formas (las activaciones de las capas de abajo hacen falta como input de las de arriba, congeladas o no), así que lo único que varía entre épocas es el coste del backward: con `K` bajo al principio, esas épocas son más baratas.

Entrenamos sobre `train_loader_nli` (sin los ejemplos traducidos) para comparar contra `model_nli` de forma aislada, igual que se hizo con el DAPT.

```python
# model_nli_swa ya no se usa; lo liberamos antes de este nuevo entrenamiento
import gc
if 'model_nli_swa' in globals():
    del model_nli_swa
gc.collect()
torch.cuda.empty_cache()

model_nli_gradual = NLIModel(model_name_nli, dropout=0.4, pooling='mean').float().to(device)

for name, param in model_nli_gradual.bert.named_parameters():
    param.requires_grad = name.startswith('encoder.layer.')
for param in model_nli_gradual.classifier.parameters():
    param.requires_grad = True

def set_unfrozen_layers(model, k):
    for name, param in model.bert.named_parameters():
        if name.startswith('encoder.layer.'):
            layer_idx = int(name.split('.')[2])
            param.requires_grad = layer_idx >= (model.bert.config.num_hidden_layers - k)

# K: 0 -> 2 -> 4 -> 6 -> 8 en 5 epochs. No llega a K=12 (encoder completo): ese intento disparo un
# reset del driver de NVIDIA (TDR) por el coste del backward completo; ver celda de arriba.
unfreeze_schedule = [0, 2, 4, 6, 8]

# el optimizer se construye una sola vez sobre TODOS los parametros del encoder + classifier (no solo
# los descongelados ahora mismo): AdamW se salta automaticamente cualquier parametro cuyo .grad sea
# None (los que siguen congelados esa epoca), asi que no hace falta reconstruirlo entre epocas
gradual_params = [p for name, p in model_nli_gradual.bert.named_parameters()
                   if name.startswith('encoder.layer.')] + list(model_nli_gradual.classifier.parameters())
optimizer_nli_gradual = torch.optim.AdamW(gradual_params, lr=2e-5, weight_decay=0.1)

epochs_nli_gradual = 5
patience_nli_gradual = 3
total_steps_nli_gradual = len(train_loader_nli) * epochs_nli_gradual
scheduler_nli_gradual = get_linear_schedule_with_warmup(
    optimizer_nli_gradual,
    num_warmup_steps=int(0.1 * total_steps_nli_gradual),
    num_training_steps=total_steps_nli_gradual)
```

```python
best_val_loss_nli_gradual = float('inf')
nli_gradual_best_val_accuracy = None
epochs_no_improve_nli_gradual = 0

for epoch in range(epochs_nli_gradual):
    k_this_epoch = unfreeze_schedule[min(epoch, len(unfreeze_schedule) - 1)]
    set_unfrozen_layers(model_nli_gradual, k_this_epoch)
    trainable_now = sum(p.numel() for p in model_nli_gradual.parameters() if p.requires_grad)

    model_nli_gradual.train()
    train_loss, train_correct, train_total = 0.0, 0, 0
    for batch in train_loader_nli:
        optimizer_nli_gradual.zero_grad()
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        token_type_ids = batch['token_type_ids'].to(device)
        batch_labels = batch['label'].to(device)

        logits = model_nli_gradual(input_ids, attention_mask, token_type_ids)
        loss = criterion(logits, batch_labels)
        loss.backward()
        torch.nn.utils.clip_grad_norm_(model_nli_gradual.parameters(), max_norm=1.0)
        optimizer_nli_gradual.step()
        scheduler_nli_gradual.step()

        train_loss += loss.item() * batch_labels.size(0)
        train_correct += (logits.argmax(dim=-1) == batch_labels).sum().item()
        train_total += batch_labels.size(0)

    model_nli_gradual.eval()
    val_loss, val_correct, val_total = 0.0, 0, 0
    with torch.no_grad():
        for batch in val_loader_nli:
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            token_type_ids = batch['token_type_ids'].to(device)
            batch_labels = batch['label'].to(device)

            logits = model_nli_gradual(input_ids, attention_mask, token_type_ids)
            loss = criterion(logits, batch_labels)

            val_loss += loss.item() * batch_labels.size(0)
            val_correct += (logits.argmax(dim=-1) == batch_labels).sum().item()
            val_total += batch_labels.size(0)

    nli_gradual_val_accuracy = val_correct / val_total
    avg_val_loss_nli_gradual = val_loss / val_total
    print(f'Epoch {epoch+1}/{epochs_nli_gradual} (K={k_this_epoch}, {trainable_now:,} params entrenables) - '
          f'loss: {train_loss/train_total:.4f} - accuracy: {train_correct/train_total:.4f} - '
          f'val_loss: {avg_val_loss_nli_gradual:.4f} - val_accuracy: {nli_gradual_val_accuracy:.4f}')

    torch.save(model_nli_gradual.state_dict(), os.path.join(checkpoint_dir, 'nli_gradual_last.pt'))

    if avg_val_loss_nli_gradual < best_val_loss_nli_gradual:
        best_val_loss_nli_gradual = avg_val_loss_nli_gradual
        nli_gradual_best_val_accuracy = nli_gradual_val_accuracy
        epochs_no_improve_nli_gradual = 0
        torch.save(model_nli_gradual.state_dict(), os.path.join(checkpoint_dir, 'nli_gradual_best.pt'))
        print(f'  -> nuevo mejor val_loss ({best_val_loss_nli_gradual:.4f}), guardado en nli_gradual_best.pt')
    else:
        epochs_no_improve_nli_gradual += 1
        print(f'  -> val_loss sin mejora desde hace {epochs_no_improve_nli_gradual} epoch(s) (mejor: {best_val_loss_nli_gradual:.4f})')
        if epochs_no_improve_nli_gradual >= patience_nli_gradual:
            print(f'Early stopping: sin mejora en val_loss durante {patience_nli_gradual} epochs consecutivos.')
            break
```

Recargamos el mejor checkpoint (`nli_gradual_best.pt`) y comparamos contra `lang_acc` (el `K=3` estático desde la época 1, sin augmentation) — el punto de interés es si esta vía (cambiar el procedimiento de entrenamiento) mueve idiomas distintos a los que movió la traducción o el DAPT.

```python
model_nli_gradual.load_state_dict(torch.load(os.path.join(checkpoint_dir, 'nli_gradual_best.pt')))
model_nli_gradual.eval()

val_preds_nli_gradual, val_true_nli_gradual = [], []
with torch.no_grad():
    for batch in val_loader_nli:
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        token_type_ids = batch['token_type_ids'].to(device)
        batch_labels = batch['label'].to(device)

        logits = model_nli_gradual(input_ids, attention_mask, token_type_ids)
        val_preds_nli_gradual.extend(logits.argmax(dim=-1).cpu().tolist())
        val_true_nli_gradual.extend(batch_labels.cpu().tolist())

lang_df_gradual = pd.DataFrame({'language': val_lang_nli, 'true': val_true_nli_gradual, 'pred': val_preds_nli_gradual})
lang_df_gradual['correct'] = lang_df_gradual['true'] == lang_df_gradual['pred']
lang_acc_gradual = lang_df_gradual.groupby('language')['correct'].agg(accuracy='mean', n='size')
overall_acc_gradual = lang_df_gradual['correct'].mean()

comparison_gradual = pd.DataFrame({
    'K3_estatico_desde_epoca1': lang_acc['accuracy'],
    'gradual_unfreezing': lang_acc_gradual['accuracy'],
})
comparison_gradual.loc['GLOBAL'] = [overall_acc_nli, overall_acc_gradual]
comparison_gradual['delta'] = comparison_gradual['gradual_unfreezing'] - comparison_gradual['K3_estatico_desde_epoca1']
comparison_gradual.sort_values('delta', ascending=False)
```

```python
plot_langs = ['Russian', 'Thai', 'Turkish', 'GLOBAL']
plot_df = comparison_gradual.loc[plot_langs]

x = np.arange(len(plot_langs))
width = 0.35

fig, ax = plt.subplots(figsize=(8, 5))
bars_1 = ax.bar(x - width / 2, plot_df['K3_estatico_desde_epoca1'], width,
                 label='K=3 estatico desde epoca 1', color='#2a78d6', zorder=3)
bars_2 = ax.bar(x + width / 2, plot_df['gradual_unfreezing'], width,
                 label='Gradual unfreezing (K: 0->2->4->6->8)', color='#d68a2a', zorder=3)

ax.set_xticks(x)
ax.set_xticklabels(plot_langs)
ax.set_ylabel('Accuracy (val, mDeBERTa, sin augmentation)')
ax.set_title('K=3 estatico vs descongelamiento gradual (hasta K=8)')
ax.set_ylim(0, 1)
ax.grid(axis='y', color='#e1e0d9', linewidth=0.8, zorder=0)
ax.set_axisbelow(True)
for spine in ['top', 'right']:
    ax.spines[spine].set_visible(False)
ax.legend(frameon=False, loc='upper center', bbox_to_anchor=(0.5, -0.12), ncol=2)

for bars in (bars_1, bars_2):
    for bar in bars:
        ax.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.01, f'{bar.get_height():.1%}',
                ha='center', fontsize=9, color='#0b0b0b')

plt.tight_layout()
plt.show()
```

**Resultado: la misma mejora que con `K` tope 3, ni más ni menos — la capacidad extra no ayuda, y confirma el mismo techo de overfitting ya visto con `K=6` estático.** El mejor checkpoint no fue el de la última época (`K=8`), sino el de la **época 3 (`K=4`, val_loss 0.3401)** — las épocas 4 y 5 (`K=6` y `K=8`) siguieron subiendo train_accuracy (88.3% → 90.3%) mientras la val_loss empeoraba, la misma señal de overfitting que ya vimos con el `K=6` estático. El early-stopping-por-mejor-checkpoint hace su trabajo y se queda con la época 3.

La accuracy global da **88.08%** — prácticamente idéntica a la del intento anterior con tope `K=3` (también 88.08%), a pesar de que aquí el modelo llegó a tener hasta 56.7M parámetros entrenables (`K=8`) frente a los 21.3M del tope anterior. Confirma, con un experimento distinto, lo que ya se había visto con `K=6` estático: **más capacidad entrenable no mueve el techo de esta tarea con este dataset** — el techo está en otro sitio (cantidad/calidad de los datos, no en cuánto del encoder se entrena). Por idioma, ganancias parecidas a las del intento anterior en francés (+3.8pts), ruso (+2.4), turco (+1.9); pero esta vez con más colateral negativo — **tailandés baja a 79.7%** (-1.4pts, la primera vez que una de estas estrategias empeora tailandés en vez de dejarlo plano) y **vietnamita cae 4.1pts**, coherente con que las épocas finales (`K=6`, `K=8`) ya estaban sobreajustando antes de que el checkpoint seleccionado las descartara.

**Conclusión práctica:** ampliar la rampa de `K=3` a `K=8` no aporta nada sobre el resultado ya obtenido — confirma más que amplía el hallazgo anterior. La ganancia real de esta técnica viene de introducir el descongelamiento *progresivo* en sí (evitar que el classifier nuevo mande gradientes ruidosos hacia el encoder desde la época 1), no de cuánta capacidad se acaba liberando. Ir más allá de `K=4`-`K=6` en este dataset small-data es puramente coste (tiempo de entrenamiento, y en este caso también el riesgo de reset de driver al llegar a `K=12`) sin beneficio.
